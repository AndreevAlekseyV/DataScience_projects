Вы сотрудник Отдела Data Science популярного музыкального стримингового сервиса "МиФаСоль". Сервис расширяет работу с новыми артистами и музыкантами, в связи с чем возникла задача -- правильно классифицировать новые музыкальные треки, чтобы улучшить работу рекомендательной системы. Ваши коллеги из отдела работы со звуком подготовили датасет, в котором собраны некоторые характеристики музыкальных произведений и их жанры. Ваша задача -- разработать модель, позволяющую классифицировать музыкальные произведения по жанрам.

Описание полей данных:

- instance_id - уникальный идентификатор трека
- track_name - название трека
- acousticness - акустичность
- danceability - танцевальность
- duration_ms -продолжительность в милисекундах
- energy - энергичность
- instrumentalness - инструментальность
- key - базовый ключ (нота) произведения
- liveness - привлекательность
- loudness - громкость
- mode - указывает на модальность (мажорную или минорную) трека
- speechiness - выразительность
- tempo - темп
- obtained_date - дата загрузки в сервис
- valence - привлекательность произведения для пользователей сервиса
- music_genre - музыкальный жанр

В качестве исходных данных представлены два файла *kaggle_music_genre_train.csv* и *kaggle_music_genre_test.csv* с обучающими и тестовыми данными. В тестовом файле *kaggle_music_genre_test.csv* столбец *music_genre* с целевым показателем отсутствует.

### Обработка столбцов

Обработку столбцов будем выполнять следующим образом. Для обучающего датафрейма строки с пропусками будут удалены, так как их колличество не велико. Для тестового датафрейма такие строки заполним медианными значениями, которые будет вычисляться на основе данных обоих датафреймов. Аномальные значения обоих датафреймов будут заменены медианными значениями.

#### Обработка столбца *key*

Выведем на экран количество каждой категории столбца.

```python
pd.concat([df_train['key'], df_test['key']]).value_counts()
```

Заполним пропуска в столбце *key* тестового датафрейма значениями *G* как наиболее часто встречающимся. В тренировочном датафрейме пропуски по данному столбцу удалим.

```python
df_test['key'] = df_test['key'].fillna('G')
```

```python
df_train = df_train.dropna(subset=['key'])
```

#### Обработка столбца *mode*

Так как категории в столбце *mode* можно соотнести по высоте звука, то его обработка ("ручная кодировка") будет выполнена исходя из следующих соображений:
1. Значениям *Major* будет присвоино число 1;
2. Значениям *Миnor* будет присвоино число 0;
3. Пропущенным значениям тестового датафрейма будет присвоена доля наиболее встречающегося значения "Major", как вероятность получения данного значения.

При этом, для тренировочной выборки пропуска будут удалены.

```python
mode_mean = pd.concat([df_train['mode'], df_test['mode']]).value_counts(normalize=True)
mode_mean
```

```python
df_test['mode'] = df_test['mode'].fillna(mode_mean[0])
```

```python
df_train['mode'] = df_train['mode'].replace('Major',1)
df_train['mode'] = df_train['mode'].replace('Minor',0)
```

```python
df_test['mode'] = df_test['mode'].replace('Major',1)
df_test['mode'] = df_test['mode'].replace('Minor',0)
```

```python
df_train = df_train.dropna(subset=['mode'])
```

#### Обработка столбца *duration_ms*

При выполнении анализа в пункте 2 мы выявили аномальные значения в данном столбце. Для их обработки выполним замену данных анамалий медианным значением.

```python
duration_concat_median = pd.concat([df_train.query('duration_ms>0'), df_test.query('duration_ms>0')])['duration_ms'].median()
duration_concat_median
```

```python
df_test.loc[df_test['duration_ms']<=0,'duration_ms'] = duration_concat_median
```

```python
df_train.loc[df_train['duration_ms']<=0,'duration_ms'] = duration_concat_median
```

#### Обработка столбца *tempo*

Выполним обработку пропусков в данном столбце путём их заполнения медианным значением в тестовом датафрейме и удалением строк в тренировочном.

```python
tempo_concat_median = pd.concat([df_train['tempo'], df_test['tempo']]).median()
tempo_concat_median
```

```python
df_test['tempo'] = df_test['tempo'].fillna(df_test['tempo'].median())
```

```python
df_train = df_train.dropna(subset=['tempo'])
```

### Создание модели *model_cb*

1. Список X - с перечнем названий столбцов с обучающими признаками;
2. Список cat_features - с перечнем названий столбцов с категориальными обучающими признаками;

```python
X = ['acousticness', 'danceability', 'duration_ms', 'energy',
     'instrumentalness', 'key', 'liveness', 'loudness', 'mode',
     'speechiness', 'tempo', 'obtained_date', 'valence']

cat_features = ['obtained_date']

y = ['music_genre']
```

Создадим список гиперпараметров модели и проведём её кросс-валидацию.

```python
params = {'cat_features': cat_features,
              'eval_metric': 'TotalF1',
          'loss_function': 'MultiClass',
              'learning_rate': 0.03,
              'random_seed': 7865,
              'verbose':100}
```

```python
cv_data = cv(
    params = params,
    pool = train_data,
    fold_count=5,
    stratified=True,
    early_stopping_rounds=200
)
```

Для оценки эффективности работы модели выведем на экран строку, при которой было получено максимальное значение *TotalF1*.

```python
cv_data[cv_data['test-TotalF1-mean'] == cv_data['test-TotalF1-mean'].max()]
```

Максимальное значение *TotalF1* получено на итеррации, близкой к 1000, что является хорошим показателем работы модели.

Таким образом выполним обучение создадим модель *model_cb* с подобранными гиперпараметрами и обучим её на тренировочных данных.

```python
model_cb = CatBoostClassifier(**params)
```

```python
model_cb.fit(train_data)
```

Определим метрику f1_micro для полученной модели. Оценку будем выполнять кросс-валидацией с помощью инструмента GridSearchCV задав только один принятый в ней гиперпараметр. 

```python
param_grid = {'depth': [6]}
```

```python
grid_search = GridSearchCV(estimator=model_cb, param_grid=param_grid, scoring=f1_mic)
```

```python
grid_search.fit(df_train_cb[X], df_train_cb[y], cat_features=cat_features)
```

```python
grid_search.best_score_
```

Полученное значение f1_micro соотавляет 0,4846, что выше значений моделей *model_rf* случайного леса и *model_tree* дерева решений.
